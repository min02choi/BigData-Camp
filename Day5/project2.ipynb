```python
import time

from bs4 import BeautifulSoup as BS
import requests

result = {}

for page in range(1, 6):
    res = requests.get(
        f"https://github.com/search?o=desc&p={page}&q=python&s=stars&type=Repositories"
    )
    soup = BS(res.text, "html.parser")

    for child in soup.select(
        "body > div.logged-out.env-production.page-responsive > div.application-main > main > div > div.col-12.col-md-9.float-left.px-2.pt-3.pt-md-0.codesearch-results > div > ul > li"
    ):
        result[
            child.select_one("div.mt-n1.flex-auto > div.d-flex > div > a").text
        ] = child.select_one(
            "div.mt-n1.flex-auto > div:nth-child(3) > div.d-flex.flex-wrap.text-small.color-fg-muted > div:nth-child(1) > a"
        ).text.strip()

    time.sleep(2)
```

```python
len(result)
```

```python
# tqdm 사용
import time

from bs4 import BeautifulSoup as BS
from tqdm import tqdm
import requests

result = {}

for page in tqdm(range(1, 6)):
    res = requests.get(
        f"https://github.com/search?o=desc&p={page}&q=python&s=stars&type=Repositories"
    )
    soup = BS(res.text, "html.parser")

    for child in soup.select(
        "#js-pjax-container > div > div.col-12.col-md-9.float-left.px-2.pt-3.pt-md-0.codesearch-results > div > ul > li"
    ):
        result[
            child.select_one("div.mt-n1.flex-auto > div.d-flex > div > a").text
        ] = child.select_one(
            "div.mt-n1.flex-auto > div:nth-child(3) > div.d-flex.flex-wrap.text-small.color-fg-muted > div:nth-child(1) > a"
        ).text.strip()

    time.sleep(3)
```
