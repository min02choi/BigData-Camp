## 파이썬에서 HTTP 요청 보내기

```python
import requests

res = requests.get('https://api.github.com/events')
```

```python
res.url         # 요청 url 확인
```

```python
res.json()      # 응답 데이터를 JSON 형식으로 확인
```

```python
headers = {'user-agent': 'dummy/0.0.1'}
res = requests.get('https://api.github.com/events', headers=headers)
res.status_code     # 상태 코드 반환
```

```python
res.headers
```

### Exercise

```python
# 모든 프로덕트
response = requests.get('https://dummyjson.com/products')
```

```python
# 프로덕트 중 핸드폰
response = requests.get('https://dummyjson.com/products/search?q=phone')
```

```python
# 새로운 프로덕트 추가
url = 'https://dummyjson.com/products/add'
headers = { 'Content-Type': 'application/json' }
data = { 'title': 'BMW Pencil' }

response = requests.post(url, headers=headers, json=data)
```

```python
# 프로덕트 수정
url = 'https://dummyjson.com/products/1'
headers = { 'Content-Type': 'application/json' }
data = { 'title': 'iPhone Galaxy +1' }

response = requests.put(url, headers=headers, json=data)
```

```python
# 프로덕트 삭제
url = 'https://dummyjson.com/products/1'    # 아이디가 1인 프로덕트 삭제

response = requests.delete(url)
```
